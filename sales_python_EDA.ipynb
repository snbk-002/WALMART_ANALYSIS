#### Data Exploration

```python
import pandas as pd
print(pd.__version__)
```

```python
df =pd.read_csv("/content/Walmart.csv")
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'unit_price': 'unit_price($)'}, inplace=True)
df.head(2)
```

```python
df['unit_price($)'] = df['unit_price($)'].str.replace('$', '').astype('float')
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
df.duplicated().sum() ##all duplicates
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
##droping missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
## Adding a New column
df["Total Price"] = df["unit_price($)"]	* df["quantity"]
```
